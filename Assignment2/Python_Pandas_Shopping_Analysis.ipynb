## Step 3 — Handle Missing Values
### 3a. Identify missing values


```python
# Count of nulls per column
missing = df.isnull().sum()
pct     = (df.isnull().mean() * 100).round(2)
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': pct})
print(missing_df[missing_df['Missing Count'] > 0])
```

### 3b. Fill / Drop missing values

| Column | Strategy | Reason |
|--------|----------|--------|
| `price` | Fill with **median** | Numeric — median is robust to outliers |
| `quantity` | Fill with **mode** (most common) | Discrete count — mode makes business sense |
| `size` | Fill with `'Unknown'` | Categorical — explicit label is cleaner than dropping |


```python
# Fill numeric: price → median
median_price = df['price'].median()
df['price'] = df['price'].fillna(median_price)
print(f"Filled 'price' nulls with median = {median_price:.2f}")

# Fill numeric: quantity → mode
mode_qty = int(df['quantity'].mode()[0])
df['quantity'] = df['quantity'].fillna(mode_qty)
print(f"Filled 'quantity' nulls with mode = {mode_qty}")

# Fill categorical: size → 'Unknown'
df['size'] = df['size'].fillna('Unknown')
print("Filled 'size' nulls with 'Unknown'")
```

```python
# Confirm no more missing values
print("=== Missing Values After Cleaning ===")
print(df.isnull().sum())
assert df.isnull().sum().sum() == 0, "⚠️ Still has nulls!"
print("\n✅ No missing values remaining.")
```

## Step 4 — Filter Rows & Select Columns


```python
# 4a. Select specific columns
print("=== Selected columns: product, category, price, quantity ===")
df[['product', 'category', 'price', 'quantity']].head(8)
```

```python
# 4b. Filter rows where price > 500
high_value = df[df['price'] > 500]
print(f"Orders with price > ₹500: {len(high_value)}")
high_value[['order_id', 'product', 'category', 'price']].head(10)
```

```python
# 4c. Filter Electronics orders with rating >= 4
top_electronics = df[(df['category'] == 'Electronics') & (df['rating'] >= 4)]
print(f"Top-rated Electronics orders: {len(top_electronics)}")
top_electronics[['order_id', 'product', 'price', 'rating']].head(8)
```

```python
# 4d. Filter Clothing orders in size 'M' or 'L'
clothing_ml = df[(df['category'] == 'Clothing') & (df['size'].isin(['M', 'L']))]
print(f"Clothing orders in M/L: {len(clothing_ml)}")
clothing_ml[['order_id', 'product', 'size', 'price']].head(8)
```

## Step 5 — Remove Duplicates


```python
# Check duplicates
n_dups = df.duplicated().sum()
print(f"Duplicate rows found: {n_dups}")
df[df.duplicated()][['order_id','product','price','quantity']].head()
```

```python
# Drop duplicates — keep first occurrence
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
print(f"✅ Duplicates removed. New shape: {df.shape}")
```

## Step 6 — Create a Derived Column: `total_amount`
`total_amount = price × quantity`


```python
# Ensure correct dtypes first
df['price']    = pd.to_numeric(df['price'],    errors='coerce')
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce').astype(int)

# Create derived column
df['total_amount'] = (df['price'] * df['quantity']).round(2)

print("=== Sample with total_amount ===")
df[['order_id', 'product', 'price', 'quantity', 'total_amount']].head(10)
```

```python
# Quick stats on total_amount
print("=== total_amount Statistics ===")
print(df['total_amount'].describe().round(2))
print(f"\nTop revenue category:")
print(df.groupby('category')['total_amount'].sum().sort_values(ascending=False).round(2))
```

## Step 7 — Save the Cleaned Dataset


```python
output_path = 'shopping_cleaned.csv'
df.to_csv(output_path, index=False)
print(f"✅ Cleaned dataset saved → '{output_path}'")
print(f"   Final shape : {df.shape[0]} rows × {df.shape[1]} columns")
```
